# kNN classification method

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('diabetes.csv')
```

```python
data
```

```python
from sklearn import preprocessing
```

```python
features = list(zip(data['Pregnancies'],data['Glucose'],data['BloodPressure'],data['SkinThickness'],data['Insulin'],data['BMI'],data['DiabetesPedigreeFunction'],data['Age']))
features
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors = 100)
```

```python
model.fit(features,data['Outcome'])
```

```python
predicted = model.predict([[6,148,72,35,0,33.6,0.627,50]])
if(predicted == 0):
    print('The patient does not have diabetes')
else:
    print('The patient is diabetic')
```

# Decision tree method


```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
feature_cols = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']
X = data[feature_cols]
y = data.Outcome
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,filled=True, rounded=True,
                special_characters=True,feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
